
# select the x_value and y_value

```python
AS.head(2)
```

```python
x = AS.drop(columns="Attrition",axis=1)
y = AS.iloc[:,1:2].values
```

# preprocessing the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
```

# split the train and test the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# model Building with sklearn

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
```

# train the model

```python
lm.fit(x_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
y_pred = lm.predict(x_test)
```

```python
y_pred
```

# estimate the cost

```python
from sklearn.metrics import mean_squared_error,accuracy_score
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
accuracy_score = accuracy_score(y_test,y_pred)
print("the rmse value is :",rmse)
print("the r_square value is :",accuracy_score)
```

```python
print("the model accuracy value is :",accuracy_score*100)
```
